```python
import cv2
import matplotlib.pyplot as plt

# Last inn bildet
image = cv2.imread("Connect_four.png")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Parametere for hver celle
rows, cols = 6, 7
cell_height, cell_width = 42, 42  # Omtrent høyde og bredde for hver celle
start_y, start_x = 48, 26  # Startpunktet for det første området (øverste venstre hjørne)

# Iterer over kolonner og rader for å klippe ut og vise hver celle
for column in range(cols):
    for row in range(rows):
        # Beregn koordinatene til området
        top_left_y = start_y + row * cell_height
        top_left_x = start_x + column * cell_width
        bottom_right_y = top_left_y + cell_height
        bottom_right_x = top_left_x + cell_width

        # Klipp ut området
        cropped_region = image_rgb[top_left_y:bottom_right_y, top_left_x:bottom_right_x]

        # Vis området
        plt.imshow(cropped_region)
        plt.title(f"Kolonne {column}, Rad {row}")
        plt.axis("off")
        plt.show()
```
